# Data Visualisation with Seaborn

Seaborn is a Python visualisation library based on matplotlib which provides the user with a high-level interface for creating attractive and informative statistical graphics. One of the biggest ways in which it improves <i>matplotlib</i> is that it provides a means of making graphs look much more aesthetically pleasing via simple functions.

I am going to provide an overview of how <i>seaborn</i> achieves this which will hopefully help you to understand how to create clear, attractive plots. This notebook will be sub-divided into two different sections:

- In-built themes
- Color palettes

## In-Built Themes:

When creating plots, it is desirable for them to be drawn in a manner that is aesthetically pleasing. Visualisations often attempt to convey quantitative insights to the reader, therefore it is important for them to be eye-catching and easy to comprehend.

While <i>matplotlib</i> is highly customisable when it comes to how data is visually represented, a decent knowledge of the library is required to know exactly what tweaks to make in order to create an attractive figure. The <i>seaborn</i> library tackles this problem by offering a high-level interface for controlling the appearance of <i>matplotlib</i> figures as well as a number of customised themes.

Below we can see the default appearance of a simple <i>matplotlib</i> graph:

```python
#Always execute before writing matplotlib code in Jupyter notebook cells
%matplotlib inline
```

```python
import matplotlib.pyplot as plt

plt.plot([0,1,2,3,4,5,6,7,8], [0,1,2,3,4,4,3,2,1], color='r')
plt.plot([0,1,2,3,4,5,6,7,8], [0,4,3,2,1,1,2,3,4], color='b')
plt.plot([0,1,2,3,4,5,6,7,8], [0,2,2,1,3,2,2,1,0], color='g')
plt.axis([0,8,0,5])

plt.show()
```

Let's see how the appearance of the above graph will change when we use <i>seaborn</i>. First, we must import the <i>seaborn</i> package:

```python
import seaborn as sns
```

Now, let's create the same graph again:

```python
plt.plot([0,1,2,3,4,5,6,7,8], [0,1,2,3,4,4,3,2,1], color='r')
plt.plot([0,1,2,3,4,5,6,7,8], [0,4,3,2,1,1,2,3,4], color='b')
plt.plot([0,1,2,3,4,5,6,7,8], [0,2,2,1,3,2,2,1,0], color='g')
plt.axis([0,8,0,5])

plt.show()
```

The <i>seaborn</i> graph is noticeably more aesthetically pleasing. The graph size is bigger and the lines are by default plotted against a light grey background, with white grid lines giving the graph better structure.

<i>Seaborn</i> offers a pair of functions for manipulating the aesthetic style of the plot:

- <i>axes_style()</i>
- <i>set_style()</i>

As well as this, it offers a pair of functions which relate to scaling different parts of the figure so that it can be easily incorporated into different contexts:

- <i>plotting_context()</i>
- <i>set_context()</i>

For the above two pairs, the first function returns a dictionary of parameters while the second sets the matplotlib defaults.

<i>Seaborn</i> has five preset themes which can be passed in as parameters in the <i>set_style()</i> function. These are:

- <b>darkgrid:</b> Grey background with white grid lines. This is the default theme for plots.


```python
plt.plot([1,2,3,4], [1,7,3,5])
sns.set_style("darkgrid")

plt.show()
```

- <b>whitegrid:</b> White background with grey grid lines.

```python
plt.plot([1,2,3,4],[7,2,5,4], 'gs')
plt.axis([0,5,0,8])
sns.set_style("whitegrid")

plt.show()
```

- <b>dark:</b> Grey background with no grid lines.

```python
plt.plot([1,2,3,4],[4,3,2,4])
plt.plot([1,2,3,4],[4,3,2,4], 'ro')
plt.axis([0,5,0,5])
sns.set_style("dark")

plt.show()
```

- <b>white:</b> White background with no grid lines.

```python
plt.bar([1,2,3,4],[1,2,3,4])
plt.axis([0.5,5,0,5])
sns.set_style("white")

plt.show()
```

- <b>ticks:</b> White background with no grid lines and added ticks at each interval on the x and y axis.

```python
plt.scatter([1,2,3,4],[1,2,3,4])
sns.set_style("ticks")

plt.show()
```

Since both <i>white</i> and <i>ticks</i> themes have plain white backgrounds (and so generally blend in with the colour of the page they are being displayed on), these themes can often look better with the top and right axes' spines removed. This is an example of something that is impossible to do using <i>matplotlib</i> parameters but which one can easily do using seaborn's <i>despine()</i> function:

```python
plt.plot([1,2,3,4],[1,2,3,4], linewidth=4)
plt.plot([1,2,3,4],[3.5,2.5,1.5,3], linewidth=4, color='r')
sns.set_style("ticks")

# Now remove top and right axes' spines
sns.despine()

plt.show()
```

The other <i>seaborn</i> function mentioned above which relates to the plot's aesthetics is <i>axes_style()</i>. This function returns a dictionary of parameters that describe all of the specific style settings for a given theme. 

If the function is called with no parameters, it returns the stylistic details of the current theme; alternatively, you can pass in as a parameter the name of a specific theme:

```python
sns.axes_style("darkgrid")
```

You can override the default values for the parameters of these themes using the <i>set_style()</i> function. Two parameters are passed in: the name of the theme you want to edit and a dictionary of the parameter(s) you want to change along with the value(s) you want them to be changed to:

```python
plt.plot([1,2,3,4],[1,2,3,1.5], color="blue")
plt.xlabel("x-axis", fontsize=18)
plt.ylabel("y-axis", fontsize=18)
plt.axis([1,4,1,4])

sns.set_style("darkgrid", {'axes.facecolor': 'orange', 'axes.labelcolor': 'grey'})

plt.show()
```

> <b>Note:</b> Changing the above parameters is not permanent; the changes you specify only apply to the current plot. This can be proven by calling the <i>axes_style()</i> function for the theme and noting that its default parameters have not been changed:

```python
sns.axes_style("darkgrid")
```

As mentioned earlier, <i>seaborn</i> also offers two functions which relate to the context of the plot. They are similar to the style functions in the sense that one is used for setting and the other is used for retrieval of details.

The function for setting a different context for the plot is <i>set_context()</i>, the purpose of which is to change the size of the plot depending on the context in which you intend to use it. There are four preset contexts which can be passed in as parameters in this function, namely <i>paper, notebook, talk,</i> and <i>poster</i>.

The default context in all plots so far has been <i>notebook</i>. Let's take a quick look at how the others look:

- <i>Paper:</i>

```python
plt.plot([0,1,2,3,4],[1,3,6,3,2], 'b--', [0,1,2,3,4],[3,4,5,4,1],'r-')
plt.axis([0,4,0,7])

#Sets style and context to default
sns.set()

sns.set_context("paper")

plt.show()
```

- <i>Talk:</i>

```python
plt.plot([0,1,2,3,4],[1,3,6,3,2], 'b--', [0,1,2,3,4],[3,4,5,4,1],'r-')
plt.axis([0,4,0,7])
sns.set_context("talk")

plt.show()
```

- <i>Poster:</i>

```python
plt.plot([0,1,2,3,4],[1,3,6,3,2], 'b--', [0,1,2,3,4],[3,4,5,4,1],'r-')
plt.axis([0,4,0,7])
sns.set_context("poster")

plt.show()
```

The other context function is <i>plotting_context()</i>, which returns a dictionary of parameters that describe a given context. As with <i>axes_style()</i>, calling the function with no parameters returns the details of the current context, while passing in a context name returns details of that particular context:

```python
sns.plotting_context("poster")
```

We can also edit the values of the parameters in a given context via the setting function <i>set_context()</i> like we did with <i>set_style()</i>. 

> <b>Note:</b> The <i>set_context()</i> function takes in 3 arguments which have the following default values:

> <i>set_context(context=None, font_scale=1, rc=None)</i> 

> If, when changing the parameter values for a given context, we are not passing in any values for <i>font_scale</i> (i.e. we are not overriding its default value), we must write <i><b>rc=</b>{--dictionary of parameters--}</i> when calling the function. This is because <i>set_context()</i> takes in a set number of arguments in a specific order (known as <i>positional arguments</i>), so it will automatically assume that the second parameter is <i>font_scale</i> unless specified otherwise.

```python
plt.plot([0,1,2,3,4],[1,3,6,3,2], 'b--', [0,1,2,3,4],[3,4,5,4,1],'r-')
plt.axis([0,4,0,7])
sns.set_context("poster", rc={'lines.linewidth': 4.8000000000000003, 'xtick.labelsize': 22.0})

plt.show()
```

As you can see, the size of the tick labels on the x-axis has increased along with the thickness of the lines. Note again that editing a context's parameter values is not permanent, which you can observe when you call the <i>plotting_context()</i> function:

```python
sns.plotting_context("poster")
```

To reset both the aesthetic and context parameters to their default values, use seaborn's <i>set()</i> function.

```python
sns.set()
```

## Color Palettes:

Color is an integral aspect of visualisations as it has the power to convey <i>patterns</i> in the data that it is representing.

One of the key functions we are going to come across in this section is <i>color_palette()</i> which provides an interface to many of the ways of generating colors in <i>seaborn</i>. This function will accept the following as arguments:

- The name of any seaborn palette
- The name of any matplotlib color map (except <i>jet</i>)
- A list of colors specified in any valid matplotlib format, namely RGB tuples, hex color codes, or HTML color names

Alternatively, you can call the function with no arguments which will return the current default color cycle. The format of the return value for <i>color_palette()</i> is a list of RGB tuples.

Another useful function is <i>set_palette()</i> which sets the default color cycle for all plots and takes the same arguments as <i>color_palette()</i>.

In order to decide which color palette or colormap is the best for a particular set of data, we need to know the characteristics of that data. Color palettes can be split into 3 distinct sub-categories:

- Qualitative
- Sequential
- Diverging

### <i>Qualitative</i>:
These types of palettes are most suitable when trying to distinguish discrete chunks of data which don't have any particular order. 

The default color cycle when you import seaborn is a set of six colors which are a more aesthetically pleasing version of the standard matplotlib color cycle. We can display them and any other color palettes using seaborn's <i>palplot()</i> function which plots the values in a color palette as a horizontal array

```python
sns.palplot(sns.color_palette())
```

The default theme can be varied to create palettes with a different appearance by using any of the following six parameters:

- <b>muted:</b>

```python
sns.palplot(sns.color_palette("muted"))
```

- <b>pastel:</b>

```python
sns.palplot(sns.color_palette("pastel"))
```

- <b>bright:</b>

```python
sns.palplot(sns.color_palette("bright"))
```

- <b>dark:</b>

```python
sns.palplot(sns.color_palette("dark"))
```

- <b>colorblind:</b>

```python
sns.palplot(sns.color_palette("colorblind"))
```

- <b>deep:</b>

```python
sns.palplot(sns.color_palette("deep"))
```

You can also pass in a second parameter in the <i>color_palette()</i> function which specifies the number of colors you want to be displayed from the palette. The default number for named palettes such as <i>pastel</i> is six; if you specify a number which is larger than the number of colors the palette contains, it will cycle. See below for a demonstration of this using the <i>colorblind</i> palette which only contains six colors:

```python
sns.palplot(sns.color_palette("colorblind", 12))
```

As you can see, it cycles once it displays its six colors once.

Another parameter that <i>color_palette()</i> accepts is a list of colors. Perhaps you just want a specific set of colors to represent your data, in which case you can do so using lists:

```python
colors = ['red','yellow','green','blue','orange','purple']
sns.palplot(sns.color_palette(colors))
```

If the number of categories you need to distinguish within your data exceeds the number of colors in a particular palette, a very useful way of overcoming this problem is using the <i>hls</i> color space. This is a transformation of RGB values which can be passed into the <i>color_palette()</i> function along with the number of distinct categories of color you want.

```python
sns.palplot(sns.color_palette("hls", 20))
```

There also exists a function called <i>hls_palette()</i> which gives you the added ability to control the lightness ( <i>l</i> ) and saturation ( <i>s</i> ) of the colors). These are represented as floats between 0 and 1.

```python
sns.palplot(sns.hls_palette(15, l=0.4, s=0.9))
```

There is an interesting function in the seaborn library called <i>choose_colorbrewer_palette()</i> which allows the user to choose palettes from the Color Brewer library (http://colorbrewer2.org/). The function takes in one mandatory argument which is the type of data you wish to visualise (one of <i>qualitative, sequential,</i> or <i>diverging</i>):

```python
sns.choose_colorbrewer_palette("qualitative")
```

### <i>Sequential:</i>

Sequential color palettes are used when you want to visualise data that ranges from low or uninteresting values to high or interesting ones. The Color Brewer library has a number of good sequential palettes which are names after the dominant color of the palette, for example:

```python
sns.palplot(sns.color_palette("Reds"))
```

If you want the darker colours to be displayed first rather than last, add a <i>_r</i> suffix to the end of the palette name, for example:

```python
sns.palplot(sns.color_palette("Greens_r"))
```

### <i>Diverging:</i>

Diverging color palettes are best suited to data sets which contain interesting low and high values. The characterisics of a diverging palette are as follows:

- Two separate, subtle hue shifts which start from distinct hues on both ends of the palette
- A well-defined mid-point where both hues meet in an under-emphasised color
- Starting values of similar brightness and saturation on each end of the palette

The Color Brewer library has a number of diverging color maps which are of interest, such as:

```python
sns.palplot(sns.color_palette("RdBu_r", 11))
```

Note in the above example all of the aforementioned characteristics of diverging color palettes: two separate colours (red and blue) at either end of the palette which start with the same level of brightness and saturation and then meet at a well-defined mid-point in an under-emphasised color (white).

If you want to create a custom colormap, you can do so using the <i>diverging_palette()</i> function. This function's two mandatory arguments are the two hues of the palette in degree format; it also has some optional arguments, such as the saturation and lightness value for the hues at either extreme of the palette.

```python
sns.palplot(sns.diverging_palette(122,321))
```

## Applying Style, Context, and Palettes to a Plot:

To show how you would apply all the ideas we have covered so far in a single plot, let's create a barchart as an example:

```python
#Make necessary imports
import seaborn as sns
import matplotlib.pyplot as plt

#Set plot style
sns.set_style("ticks")

#Set plot context
sns.set_context("talk")

#Create bar chart (which takes 'palette' as a parameter)
sns.barplot(x=[1,2,3,4], y=[4,7,2,9], palette="pastel")

#Define range of x and y axes
plt.axis([-1,4,0,10])

#Display bar chart
plt.show()
```
